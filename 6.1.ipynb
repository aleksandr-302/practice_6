```python
number=int(input("введите количество списков в списке "))
t=[]
for i in range(0,number,1):
  spisok=input("введите список ")
  spisok=[int(c) for c in spisok]
 
  t.append(spisok)
  

def nested_sum(t1):
  sm = 0
 
  for el in t1:
   
    sm += sum(el)
  return sm

print("сумма равна ",nested_sum(t))
```
